```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Load the data
data = pd.read_csv('annual_temp.csv')
data.head()
```

```python
X = data['Year'].values.reshape(-1, 1)
y = data['Mean'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

def plot_results(X, y, model, title):
    plt.scatter(X, y, color='red')
    plt.plot(X, model.predict(X), color='blue')
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Mean Temperature')
    plt.show()


def print_accuracy(y_true, y_pred, model_name):
    print(f"{model_name} R^2 Score: {r2_score(y_true, y_pred):.4f}")
    print(f"{model_name} Mean Squared Error: {mean_squared_error(y_true, y_pred):.4f}\n")


linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear_train = linear_model.predict(X_train)
y_pred_linear_test = linear_model.predict(X_test)
plot_results(X_train, y_train, linear_model, 'Linear Regression (Training)')
plot_results(X_test, y_test, linear_model, 'Linear Regression (Testing)')
print_accuracy(y_test, y_pred_linear_test, "Linear Regression")
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2, random_state=0)


poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train_poly)

y_pred_poly_train = poly_model.predict(X_train_poly)
y_pred_poly_test = poly_model.predict(X_test_poly)

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, poly_model.predict(poly.fit_transform(X_train)), color='blue')
plt.title('Polynomial Regression (Training)')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()

plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, poly_model.predict(poly.fit_transform(X_test)), color='blue')
plt.title('Polynomial Regression (Testing)')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()

print_accuracy(y_test_poly, y_pred_poly_test, "Polynomial Regression")
```

```python
# Decision Tree Regression

tree_model = DecisionTreeRegressor(random_state=0)
tree_model.fit(X_train, y_train)
y_pred_tree_train = tree_model.predict(X_train)
y_pred_tree_test = tree_model.predict(X_test)
plot_results(X_train, y_train, tree_model, 'Decision Tree Regression (Training)')
plot_results(X_test, y_test, tree_model, 'Decision Tree Regression (Testing)')
print_accuracy(y_test, y_pred_tree_test, "Decision Tree Regression")
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train_scaled = sc_X.fit_transform(X_train)
y_train_scaled = sc_y.fit_transform(y_train.reshape(-1, 1)).flatten()
X_test_scaled = sc_X.transform(X_test)

svr_model = SVR(kernel='rbf')
svr_model.fit(X_train_scaled, y_train_scaled)

y_pred_svr_train = sc_y.inverse_transform(svr_model.predict(X_train_scaled).reshape(-1, 1)).flatten()
y_pred_svr_test = sc_y.inverse_transform(svr_model.predict(X_test_scaled).reshape(-1, 1)).flatten()

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, y_pred_svr_train, color='blue')
plt.title('SVR (Training)')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()

plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred_svr_test, color='blue')
plt.title('SVR (Testing)')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.show()

def print_accuracy(y_true, y_pred, model_name):
    print(f"{model_name} R^2 Score: {r2_score(y_true, y_pred):.4f}")
    print(f"{model_name} Mean Squared Error: {mean_squared_error(y_true, y_pred):.4f}\n")

print_accuracy(y_test, y_pred_svr_test, "Support Vector Regression")
```

```python
# Random Forset Regressison

forest_model = RandomForestRegressor(n_estimators=10, random_state=0)
forest_model.fit(X_train, y_train)
y_pred_forest_train = forest_model.predict(X_train)
y_pred_forest_test = forest_model.predict(X_test)
plot_results(X_train, y_train, forest_model, 'Random Forest Regression (Training)')
plot_results(X_test, y_test, forest_model, 'Random Forest Regression (Testing)')
print_accuracy(y_test, y_pred_forest_test, "Random Forest Regression")
```
